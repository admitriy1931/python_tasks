```python
import re
```

```python
#Task1
def check_car_number(car_id):
    letters = 'АВЕКМНОРСТУХ'
    pattern = rf'^([{letters}])(\d{{3}})([{letters}]{{2}})(\d{{2,3}})$'
    match = re.match(pattern, car_id)
    if match:
        number = ''.join(match.groups()[:3])
        region = match.group(4) 
        return f"Номер {number} валиден. Регион: {region}."
    else:
        return "Номер не валиден."

print(check_car_number('А222BC96')) 
print(check_car_number('АБ22ВВ193'))  
```

```python
#Task2
def remove_duplicate_words(text):
    pattern = r'\b(\w+)\b(\s+\1)+'
    return re.sub(pattern, r'\1', text, flags=re.IGNORECASE)

some_string = 'Напишите функцию функцию, которая будет будет будет будет удалять все все все все последовательные повторы слов из из из из заданной строки строки при помощи регулярных выражений'
result = remove_duplicate_words(some_string)
print(result)
```
